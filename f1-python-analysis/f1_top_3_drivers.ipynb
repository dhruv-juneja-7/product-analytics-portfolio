```python
wins = df[df['positionOrder']==1]
```

```python
races = pd.read_csv('races.csv')
```

```python

print(df.shape, wins.shape)
```

```python
merged = pd.merge(wins, races[['raceId','year']], how='inner',on='raceId')
```

```python
merged.head()
```

```python
wins_per_season = merged.groupby(['year','driverId'])['positionOrder'].count().reset_index()
```

```python
wins_per_season.head()
```

```python
wins_per_season.columns = ['year', 'driverId', 'wins']
```

```python
wins_per_season.head()
```

```python
wins_per_season['rank'] = wins_per_season.groupby('year')['wins'].rank(method='dense', ascending=False)
```

```python
wins_per_season.head()
```

```python
top3 = wins_per_season[wins_per_season['rank']<=3].sort_values(['year','wins'], ascending=[False,False])
```

```python
print(top3.head(20))
```

```python
drivers = pd.read_csv('drivers.csv')
```

```python
drivers['driverName'] = drivers['forename'] + ' ' + drivers['surname']
```

```python
top_3_drivers = top3.merge(drivers[['driverId', 'driverName']], on='driverId').sort_values(['year','wins'], ascending=[False,False])
```
