```python
import pandas as pd
```

```python
races = pd.read_csv('races.csv')
results = pd.read_csv('results.csv')
drivers = pd.read_csv('drivers.csv')
constructors = pd.read_csv('constructors.csv')
driver_standings = pd.read_csv('driver_standings.csv')
```

```python
merged = results[['raceId', 'driverId', 'constructorId','positionOrder']].merge(races[['raceId','year']], on='raceId',how='left')
```

```python
merged = merged.merge(drivers[['driverId','forename','surname']],on='driverId', how='left')
```

```python
merged = merged.merge(constructors[['constructorId', 'name']], on='constructorId', how='left')
```

```python
merged = merged.merge(driver_standings[['raceId','driverId','points']], on=['raceId','driverId'], how='left')
```

```python
merged['driver_name'] = merged['forename'] + ' ' + merged['surname']
```

```python
merged.head()
```

```python
merged_red_bull = merged.loc[(merged['name'] == 'Red Bull') & (merged['year'].between(2021,2024))]
```

```python
avg_position = merged_red_bull.groupby(['year','driver_name']).agg(avg_position=('positionOrder', 'mean'), points=('points','max')).reset_index()
```

```python
avg_position.sort_values(['year','avg_position'], ascending=[False, True])
```

```python
max_data = avg_position.loc[avg_position['driver_name'] == 'Max Verstappen']
```

```python
perez_data = avg_position.loc[avg_position['driver_name'] == 'Sergio Pérez']
```

```python
combined_data = max_data.merge(perez_data, on='year', how='left')
```

```python
combined_data = combined_data.rename(columns={'driver_name_x':'driver_1', 'avg_position_x':'avg_pos_driver_1', 'points_x':'points_driver_1'})
```

```python
combined_data = combined_data.rename(columns={'driver_name_y':'driver_2', 'avg_position_y':'avg_pos_driver_2', 'points_y':'points_driver_2'})
```

```python
combined_data
```

```python
combined_data['points_gap'] = combined_data['points_driver_1'] - combined_data['points_driver_2']
```

```python
combined_data['position_gap'] = combined_data['avg_pos_driver_2'] - combined_data['avg_pos_driver_1']
```

```python
combined_data
```

```python
combined_data = combined_data.round(2)
```
